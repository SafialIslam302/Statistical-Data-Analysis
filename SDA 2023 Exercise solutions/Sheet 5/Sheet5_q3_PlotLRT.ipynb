```python
import numpy as np
from matplotlib import pyplot as plt


plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Helvetica"
})
```

```python
thetaList = [-2,0,0.5,1]
n = 1
x = np.linspace(-3,3,600)
s = []
for theta in thetaList:
    y = np.exp(n*(theta-x))
    y[x<=theta] = 1
    plt.plot(x,y)
    s.append('$\\theta_0 =$ ' + str(theta))


plt.xlabel('$x_{(1)}$')
plt.ylabel('$\lambda(x_1,\dots,x_n)$')
plt.title('Likelihood ratio test $\lambda(x)$')
plt.legend(s)
plt.show()
```
